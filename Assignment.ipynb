```python
(3)SEARCHING AND SORTING ALGORITHMS

#linear search to find 32
arr = [21, 10, 45, 32, 67, 5, 18]
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

result = linear_search(arr, 32)
print(result)
```

```python
#Binary search to find 45
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1

sorted_arr = sorted(arr)
result = binary_search(sorted_arr, 45)
print(result)
```

```python
#Bubble sort 
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

sorted_list = bubble_sort(arr.copy())
print(sorted_list)
```

```python
(4)LINKED LIST IMPLEMENTATION

#Defining a node
class Node:
    def __init__(self, data):
        self.data = data #data =stores the value of the node 
        self.next = None #next =stores the reference to the next node
#Linked list class    
class LinkedList:
    def __init__(self):
        self.head = None

#Inserting node at beginning
    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

#Inserting node at end
    def insert_at_end(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next

        temp.next = new_node

#Displaying all elements
    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")
```
